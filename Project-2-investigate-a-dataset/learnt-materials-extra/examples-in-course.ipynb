```python
from math import sqrt

def fibo1(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibo1(n-1) + fibo1(n-2)
def fibo2(n):
    return ((1 + sqrt(5))**n - (1 - sqrt(5))**n)/(2**n*sqrt(5))
```

```python
%timeit fibo1(20)
```

```python
%timeit fibo2(20)
```

```python
import random
```

```python
%%timeit
prize = 0
for ii in range(100):
    ##roll a die
    roll = random.randint(1,6)
    if roll % 2 == 0:
        prize += roll
    else:
        prize -= 1
```

```python
%%timeit
rolls = (random.randint(1,6) for _ in range(100))
prize = sum(roll if roll % 2 == 0 else -1 for roll in rolls)
```
